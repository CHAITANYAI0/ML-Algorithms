```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris') # Other way to get iris dataset
```

```python
df.head(3)
```

```python
X = df.iloc[:,:-1]
y = iris.target
```

```python
X,y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#Post Prunning
DTC = DecisionTreeClassifier()
```

```python
DTC.fit(X_train,y_train)
```

```python
from sklearn import tree
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(DTC, filled=True)
```

```python
DTC_HP = DecisionTreeClassifier(max_depth=2)
```

```python
DTC_HP.fit(X_train,y_train)
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(DTC_HP, filled=True)
```

```python
y_pred_HP = DTC_HP.predict(X_test)
y_pred = DTC.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
score_HP = accuracy_score(y_pred_HP, y_test)
score = accuracy_score(y_pred, y_test)

print(score_HP)
print(score)
```

```python
print(classification_report(y_pred_HP, y_test))
print(classification_report(y_pred, y_test))
```
